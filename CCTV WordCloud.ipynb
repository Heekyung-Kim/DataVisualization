```python
from bs4 import BeautifulSoup
import requests
from konlpy.tag import Twitter
from collections import Counter
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
search_word = 'CCTV'
title_list = []

def get_titles(start_num, end_num):
    while 1:
        if start_num > end_num:
            break
        print(start_num)
        
        url = 'https://search.naver.com/search.naver?where=news&sm=tab_jum&query={}&start={}'.format(search_word, start_num)
        req = requests.get(url)
        
        if req.ok:
            html = req.text
            soup = BeautifulSoup(html, 'html.parser')
            print('연결성공')
            
            # 뉴스 제목 가져오기
            titles = soup.select(
                'ul.type01 > li > dl > dt > a'
            )
            
            for title in titles:
                title_list.append(title['title'])
        start_num += 5
    print(title_list)
```

```python
def make_wordcloud(word_count):
    twitter = Twitter()
    
    sentences_tag = []
    
    # 형태소분석 -> 리스트에 넣기
    for sentence in title_list :
        morph = twitter.pos(sentence)
        sentences_tag.append(morph)
        print(morph)
        print('-'*30)
        
    print(sentences_tag)
    print('\n'*3)
    
    # 명사, 형용사만 구분해서 리스트에 넣기
    noun_adj_list = []
    
    for sentence1 in sentences_tag:
        for word, tag in sentence1:
            if tag in ['Noun', 'Adjective']:
                noun_adj_list.append(word)
                
    # 형태소별 count
    counts = Counter(noun_adj_list)
    tags = counts.most_common(word_count)
    print(tags)
    
    # word cloud 생성
    # 한글 폰트 지정
    
    wc = WordCloud(font_path = 'c:\windows\fonts\malgun.ttf')
    print(dict(tags))
    cloud = wc.generate_from_frequencies(dict(tags))
    plt.figure(figsize = (10,8))
    plt.axis('off')
    plt.imshow(cloud)
    plt.show()
    
if __name__== '__main__':
    get_titles(1,200) # 가져올 게시글 범위 지정
    
    make_wordcloud(30) # 클라우드로 만들 단어 개수 지정
    
```
